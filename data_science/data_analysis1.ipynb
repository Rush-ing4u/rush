 - Selecting rows & columns  
    - Rows (slicing)

          - ` loc` - gets us the data using named indexes

          - ` iloc`- gets us the data using numeric indexes
    - Columns (dictionary style)

          - ` df[single]` - requires name of columns 
          
          - ` df[multiple]` - requires a list of columns (`multiple` is list of named columns )


```python
# single row (record)
weather.iloc[0]
```

```python
# record 25 to 50
weather.iloc[25:50]
```

```python
# single column 
weather.iloc[1:16 , :5]
```

```python
weather.iloc[:15 , -3:]
```

```python
weather.AvgTemp  # same as weather['AvgTemp'] (object style)
```
